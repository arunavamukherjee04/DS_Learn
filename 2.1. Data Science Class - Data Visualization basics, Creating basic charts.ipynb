## Creating linechart from a list object

### Plotting a line chart in matplotlib

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]
plt.plot(x, y)
```

### Plotting a line chart from pandas object

```python
address = "Data Set/Northwind.xls"
excel_file = pd.ExcelFile(address)
excel_file.sheet_names
products = excel_file.parse('Product')
products.columns
```

```python
products
```

```python
unit_price = products['UnitPrice']
unit_price.plot()
```

```python
units_in_stock = products['UnitsInStock']
units_in_stock.plot()
```

```python
# with multiple variable
data_frame = products[['UnitPrice', 'UnitsInStock']]
data_frame.plot()
```
