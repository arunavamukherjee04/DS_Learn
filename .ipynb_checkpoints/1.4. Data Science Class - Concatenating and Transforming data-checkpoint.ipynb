```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
DF_obj = DataFrame(np.arange(49).reshape(7, 7))
DF_obj
```

```python
DF_obj_2 = DataFrame(np.arange(20).reshape(5, 4))
DF_obj_2
```

## Concating Data

```python
#concat method joins data from seperate sources into one combined data table
# axis = 1 tells python to concatenate the data frames by adding the columns
pd.concat([DF_obj, DF_obj_2], axis = 1)
```

```python
# If you don't mention axis = 1, then by default it will concatenate on columns

pd.concat([DF_obj, DF_obj_2])
```

## Transforming the data

### Adding Data


```python
series_obj = Series(np.arange(6))
series_obj.name = "series_Object"
series_obj
```

```python
joined_variable = DataFrame.join(DF_obj, series_obj)
joined_variable
```

```python
added_datatable = joined_variable.append(joined_variable, ignore_index = False)
added_datatable
```

### Sorting the data


```python
added_datatable.sort_values(by=[1], ascending=[False])
```
